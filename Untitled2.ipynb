```python
# Simple Colab: convert alternating-line .txt -> Excel (col A = digits, col B = words)
!pip install openpyxl --quiet

import pandas as pd
from google.colab import files

print("Upload a .txt file where lines alternate: digit, word, digit, word, ...")
uploaded = files.upload()

for fname, fileinfo in uploaded.items():
    print(f"\nProcessing {fname} ...")
    # fileinfo is bytes
    text = fileinfo.decode('utf-8', errors='replace')
    # split into lines and remove empty lines
    lines = [ln.strip() for ln in text.splitlines() if ln.strip() != ""]

    # If odd number of lines, append empty string so last pair still created
    if len(lines) % 2 == 1:
        lines.append("")

    # Pair up: even index -> column A, odd index -> column B
    rows = []
    for i in range(0, len(lines), 2):
        a = lines[i]
        b = lines[i+1]
        rows.append((a, b))

    # Create DataFrame with desired column order (A then B)
    df = pd.DataFrame(rows, columns=['Column A', 'Column B'])

    # Save to excel
    out_name = fname.rsplit('.', 1)[0] + '.xlsx'
    df.to_excel(out_name, index=False)
    print(f"Saved: {out_name} (rows: {len(df)})")

    # trigger download
    try:
        files.download(out_name)
    except Exception:
        print("Download link should appear in the Colab file browser on the left.")
```

```python
# Install dependencies
!pip install pandas openpyxl --quiet

import pandas as pd
from google.colab import files

# Step 1: Upload the two Excel files
print("Upload the FIRST Excel file (e.g. digits+YES_NO.xlsx)")
uploaded1 = files.upload()

print("Upload the SECOND Excel file (e.g. digits+text.xlsx)")
uploaded2 = files.upload()

# Step 2: Read both Excel files into pandas DataFrames
file1 = list(uploaded1.keys())[0]
file2 = list(uploaded2.keys())[0]

df1 = pd.read_excel(file1)
df2 = pd.read_excel(file2)

# Step 3: Merge them on 'Column A'
# This performs an inner join (only matching IDs).
# For all rows from both files, use `how="outer"`.
merged = pd.merge(df1, df2, on="Column A", how="inner")

# Step 4: Save merged file to Excel
output_filename = "merged_result.xlsx"
merged.to_excel(output_filename, index=False)

# Step 5: Download result
files.download(output_filename)

print("✅ Merged successfully! File saved as merged_result.xlsx")
```

##Data Preprocessing



```python
# Drop duplicate IDs (keeping first)
df = df.drop_duplicates(subset="Sentence", keep="first")

# Drop rows where Text or Label is missing
df = df.dropna(subset=["Sentence", "Output"])

print("✅ After removing duplicates and missing data:", df.shape)
```

```python
import re
def clean_text(text):
    text = str(text).lower()  # lowercase
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)  # remove URLs
    text = re.sub(r"@\w+", "", text)  # remove mentions (@user)
    text = re.sub(r"#", "", text)  # remove hashtag symbol but keep word
    text = re.sub(r"[^a-zA-Z0-9\s]", " ", text)  # remove special chars & emojis
    text = re.sub(r"\s+", " ", text).strip()  # remove extra spaces
    return text

df["Cleaned_Sentence"] = df["Sentence"].apply(clean_text)
```

```python
df.head()
```

```python
# Replace Sentence content with cleaned text
df["Sentence"] = df["Cleaned_Sentence"]

# Drop the Cleaned_Sentence column
df = df.drop(columns=["Cleaned_Sentence"])

# Verify result
df.head()
```

```python
# Install nltk if not installed
!pip install nltk --quiet

import nltk
from nltk.tokenize import word_tokenize
```

```python
# 🔽 Download required resources
nltk.download('punkt')        # standard tokenizer data
nltk.download('punkt_tab')    # sometimes required for newer NLTK versions
```

```python
df['Word_Tokens'] = df['Sentence'].apply(word_tokenize)
```

```python
df.head()
```

```python
def perform_stemming(word_list):
    stemmer = PorterStemmer()
    stemmed_words = [stemmer.stem(word) for word in word_list]
    return stemmed_words


df['Word_Tokens'] = df['Word_Tokens'].apply(perform_stemming)
```

```python
df.head()
```

```python
def custom_lemmatization(word_list):
    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in word_list]
    return lemmatized_words

df['Word_Tokens'] = df['Word_Tokens'].apply(custom_lemmatization)
```

```python
df.head()
```

```python
# Convert list of tokens to sentence (string)
df['Source'] = df['Word_Tokens'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)

# Verify result
df[['Sentence', 'Source']].head()
```

```python
df['Sentence'] = df['Source']
df = df.drop(columns=['Source', 'Word_Tokens'])
df.head()
```

```python
# Convert YES/NO to 1/0
df['Output'] = df['Output'].map({'YES': 1, 'NO': 0})

# Check unique values to confirm
print("Unique values after encoding:", df['Output'].unique())

# Preview
df.head()
```

```python
X=df['Sentence'].tolist()
y=df['Output'].tolist()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # You can adjust the max_features parameter
X_tfidf = tfidf_vectorizer.fit_transform(X)
```

##Data Splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.3, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
# Assuming your X_train data is already loaded and in the correct format
# Initialize SMOTE with 'auto' strategy (to automatically detect the minority class)
smote = SMOTE(sampling_strategy='auto')

# Fit SMOTE on X_train data
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
import numpy as np

# Count the occurrences of each class in the original training set
unique_classes, original_class_counts = np.unique(y_train, return_counts=True)

# Count the occurrences of each class in the resampled training set after applying SMOTE
unique_classes_resampled, resampled_class_counts = np.unique(y_train_resampled, return_counts=True)

# Display the class distribution for both the original and resampled datasets
print("Original Class Distribution:")
for cls, count in zip(unique_classes, original_class_counts):
    print(f"Class {cls}: {count} samples")

print("\nResampled Class Distribution:")
for cls, count in zip(unique_classes_resampled, resampled_class_counts):
    print(f"Class {cls}: {count} samples")
```

##Model

```python
# Assuming you have already applied SMOTE and obtained the resampled training data
# X_train_resampled and y_train_resampled

# Create a DataFrame combining the resampled features and labels
# If your data is in NumPy arrays, you can convert them to DataFrame as follows:
#df_resampled = pd.DataFrame(X_train_resampled, columns=['feature1', 'feature2', 'feature3', 'feature4'])
#df_resampled['target'] = y_train_resampled

# Plot the count plot using seaborn
plt.figure(figsize=(8, 6))
sb.countplot(x=y_train_resampled)
plt.xlabel('Is_Sarcastic', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Class Distribution after Resampling', fontsize=16)
plt.show()
plt.savefig("AfterResample.png")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Initialize the Logistic Regression model
lr = LogisticRegression(random_state=42)

# Train the model on the training data
lr.fit(X_train_resampled, y_train_resampled)

# Make predictions on the test data
y_pred = lr.predict(X_test)

# Evaluate the model
lr_accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f"Accuracy: {lr_accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(classification_rep)
```

```python
from sklearn.svm import SVC

# Initialize the Support Vector Machine (SVM) model with a linear kernel
svm_model = SVC(kernel='linear', random_state=42)

# Train the model on the training data
svm_model.fit(X_train_resampled, y_train_resampled)

# Make predictions on the test data
y_pred = svm_model.predict(X_test)

# Evaluate the model
svm_accuracy = accuracy_score(y_test, y_pred)
svm_conf_matrix = confusion_matrix(y_test, y_pred)
svm_classification_rep = classification_report(y_test, y_pred)

print(f"Accuracy: {svm_accuracy}")
print("Confusion Matrix:")
print(svm_conf_matrix)
print("Classification Report:")
print(svm_classification_rep)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize the Decision Tree classifier
dt_model = DecisionTreeClassifier(random_state=42)

# Train the model on the training data
dt_model.fit(X_train_resampled, y_train_resampled)

# Make predictions on the test data
y_pred = dt_model.predict(X_test)

# Evaluate the model
dt_accuracy = accuracy_score(y_test, y_pred)
dt_conf_matrix = confusion_matrix(y_test, y_pred)
dt_classification_rep = classification_report(y_test, y_pred)

print(f"Accuracy: {dt_accuracy}")
print("Confusion Matrix:")
print(dt_conf_matrix)
print("Classification Report:")
print(dt_classification_rep)
```

```python
from sklearn.metrics import roc_curve, auc

# Get predicted probabilities for the test set
y_probs = dt_model.predict_proba(X_test)[:, 1]

# Calculate ROC curve values
fpr, tpr, thresholds = roc_curve(y_test, y_probs)

# Calculate AUC
roc_auc = auc(fpr, tpr)
```
